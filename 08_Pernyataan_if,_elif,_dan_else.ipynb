# Pernyataan if, elif, else

<code>if</code> Pernyataan dalam Python memungkinkan kita memberi tahu komputer untuk melakukan tindakan alternatif berdasarkan serangkaian hasil tertentu.

Secara verbal, kita dapat membayangkan bahwa kita memberi tahu komputer:

"Hei jika kasus ini terjadi, lakukan beberapa tindakan"

Kita kemudian dapat memperluas pernyataan lebih lanjut dengan pernyataan <code>elif</code> dan <code>else</code>, yang memungkinkan kita untuk memberi tahu komputer:

"Hei jika kasus ini terjadi, lakukan beberapa tindakan. Jika kasus lain terjadi, lakukan tindakan lain. Jika *tidak ada* kasus di atas terjadi, lakukan tindakan ini."

    if kasus1:
        lakukan tindakan1
    elif kasus2:
        lakukan tindakan2
    else: 
        lakukan tindakan3

## Contoh Pertama

```python
if True:
    print('Benar!')
```

Mari berikan beberapa logika!

```python
x = False

if x:
    print('x adalah benar (True)')
else:
    print('Aku akan di print jika kejadian dimana x adalah tidak benar')
```

### Cabang berlebih

Mari kita dapatkan gambaran yang lebih lengkap tentang seberapa jauh <code>if</code>, <code>elif</code>, dan <code>else</code> dapat membawa kita!

Perhatikan bagaimana <code>if</code>, <code>elif</code>, dan <code>else</code> tersusun dalam kode. Ini dapat membantu kita melihat apakah <code>if</code> terkait dengan pernyataan <code>elif</code> atau <code>else</code> apa.

Kami akan memperkenalkan kembali sintaks perbandingan untuk Python.

```python
loc = 'Bank'

if loc == 'Restoran':
    print('Selamat datang di Restoran!')
elif loc == 'Bank':
    print('Selamat datang di Bank!')
else:
    print('Kamu dimana?')
```

Perhatikan bagaimana masing-masing pernyataan <code>if</code> diperiksa hingga boolean True berjalan sampai kebawah. Kita juga harus memperhatikan bahwa kita dapat memasukkan pernyataan <code>elif</code> sebanyak yang kita inginkan sebelum kita menutup dengan <code>else</code>.

Mari buat dua contoh sederhana lagi untuk pernyataan <code>if</code>, <code>elif</code>, dan <code>else</code>:

```python
person = 'Budi'

if person == 'Budi':
    print('Selamat datang Budi!')
else:
    print("WSelamat datang, siapa namamu?")
```

```python
person = 'Andre'

if person == 'Budi':
    print('Selamat datang Budi!')
elif person =='Andre':
    print('Selamat datang Andre!')
else:
    print("WSelamat datang, siapa namamu?")
```
